```python
import threading
import datetime
from flask import Flask, request, jsonify, render_template_string

app = Flask(__name__)

# --- HTML/Frontend ---
HTML_PAGE = """
<!DOCTYPE html>
<html>
<head>
    <title>Women Safety System</title>
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <style>
        body { font-family: sans-serif; background: #fff0f3; text-align: center; padding: 20px; }
        .sos-btn { 
            background: #ff4d6d; color: white; border: none; width: 150px; height: 150px; 
            border-radius: 50%; font-size: 24px; font-weight: bold; cursor: pointer;
            box-shadow: 0 5px #c9184a;
        }
        #status { margin-top: 20px; color: #590d22; font-weight: bold; }
    </style>
</head>
<body>
    <h1>🚨 Safety Alert</h1>
    <button class="sos-btn" onclick="sendSOS()">SOS</button>
    <p id="status">Ready</p>
    <script>
    function sendSOS() {
        document.getElementById("status").innerText = "Sending...";
        navigator.geolocation.getCurrentPosition(position => {
            fetch('/sos', {
                method: 'POST',
                headers: {'Content-Type': 'application/json'},
                body: JSON.stringify({lat: position.coords.latitude, lon: position.coords.longitude})
            }).then(() => { document.getElementById("status").innerText = "🚨 ALERT SENT!"; });
        });
    }
    </script>
</body>
</html>
"""

@app.route('/')
def home():
    return render_template_string(HTML_PAGE)

@app.route('/sos', methods=['POST'])
def sos():
    data = request.json
    map_url = f"http://maps.google.com/?q={data['lat']},{data['lon']}"
    print(f"\n📢 SOS RECEIVED: {map_url}\n")
    return jsonify({"status": "ok"})

# --- Start Server and Tunnel ---
def run_app():
    app.run(port=5000, debug=False, use_reloader=False)

# Start Flask in a background thread
threading.Thread(target=run_app).start()

# Start the tunnel to give you a public link
print("Wait for the link to appear below...")
!lt --port 5000
```
